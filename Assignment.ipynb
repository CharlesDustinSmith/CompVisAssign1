# 2. Basic Pixel Editing of an Image

```python
import cv2 as cv
from IPython.display import Image

# Open the image file 
input_filename = input("Enter filename (No extensions): ")
input_filename = input_filename + ".jpg"

# Load the image
img = cv.imread(input_filename)

# Set the red channel for each pixel to 0
img[:, :, 2] = 0

# Create the output filename
output_filename = input_filename.split('.jpg')[0] + "_BLUE.jpg"

# Save the modified image
cv.imwrite(output_filename, img)

# Displaying the image
Image(output_filename)
```

# 3. Convert Image to Greyscale

```python
import cv2 as cv
from IPython.display import Image

# Get input image filename from command line
input_filename = input("Enter filename (No extensions): ")
input_filename = input_filename + ".jpg"

# Load the image
img = cv.imread(input_filename)

# Convert to Grayscale using luma method
grayscale = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Create  the output filename
output_filename = input_filename.split('.jpg')[0] + "_GREY.jpg"

#Save the modified image
cv.imwrite(output_filename, grayscale)

# Displaying the image
Image(output_filename)
```

# 4. Shift Color and Clamp

```python
import cv2 as cv
from IPython.display import Image

# Get input image filename from command line
input_filename = input("\nEnter filename (No extensions): ")
input_filename = input_filename + ".jpg"

# Get input shift value from command line
r_shift = int(input("\nEnter R shift value: "))
g_shift = int(input("\nEnter G shift value: "))
b_shift = int(input("\nEnter B shift value: "))

# Load the image 
img = cv.imread(input_filename)

# Split the image into its 3 color channels
b,g,r = cv.split(img)

# Apply the shift to each channel
b = cv.add(b, b_shift)
g = cv.add(g, g_shift)
r = cv.add(r, r_shift)

# Merge the channels back together
img = cv.merge((b, g, r))

# Clam the values of each pixel to the range [0, 255]
img = cv.convertScaleAbs(img)

# Save the output image 
output_filename = input_filename.split('.jpg')[0] + "_CLAMP.jpg"
cv.imwrite(output_filename, img)

# Displaying the image
Image(output_filename)
```
